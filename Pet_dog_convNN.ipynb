```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import time
from tensorflow.keras.callbacks import TensorBoard
```

```python
MODAL_NAME = "CatsVSDogCnn{}".format(int(time.time()))
```

```python
tensorboard = TensorBoard(log_dir='logs/{}'.format(MODAL_NAME))
```

```python
DATADIR="./PetImages"
CATEGORIES = ["Dog","Cat"]

for category in CATEGORIES:
    path = os.path.join(DATADIR,category)
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
```

```python
IMG_SIZE = 50

new_array = cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
new_array
```

```python
plt.imshow(new_array,cmap="gray")
plt.show()
```

```python
traning_data = []

def creat_traning_data():
    for category in CATEGORIES:
        path = os.path.join(DATADIR,category)
        class_num = CATEGORIES.index(category)
        for img in os.listdir(path):
            try:
                print('adding')
                gray_image = cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                proccessed_image = cv2.resize(gray_image,(IMG_SIZE,IMG_SIZE))
                traning_data.append([proccessed_image,class_num])
            except Exception as e:
                print(e)
```

```python
creat_traning_data()
```

```python
import random
```

```python
random.shuffle(traning_data)
```

```python
X = []
y = []
```

```python
for features,lable in traning_data:
    X.append(features)
    y.append(lable)
    
X = np.array(X).reshape(-1,IMG_SIZE,IMG_SIZE,1)
y = np.array(y).reshape(-1)
```

```python
y
```

```python
import pickle
```

```python
pickle_out = open("X.pickle","wb")
pickle.dump(X,pickle_out)
pickle_out.close()

pickle_out = open("y.pickle","wb")
pickle.dump(y,pickle_out)
pickle_out.close()
```

```python
pickle_in = open("X.pickle","rb")
X = pickle.load(pickle_in)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Activation,Flatten,Conv2D,MaxPooling2D
```

```python
X = X/255.0
```

```python
model = Sequential()
model.add(Conv2D(64,(3,3), input_shape = X.shape[1:]))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64,(3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation("relu"))

model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss="binary_crossentropy",
             optimizer='adam',
              metrics=['accuracy']
             )

model.fit(X,y,batch_size=32,epochs=3,validation_split=0.1,callbacks=[tensorboard])
```
